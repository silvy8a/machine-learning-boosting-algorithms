Predicting diabetes

```python
import pandas as pd
#TOTAL DATA
#Reading the total_data_sel dataset
total_data_train_sel = pd.read_csv("/workspaces/machine-learning-boosting-algorithms/data/processed/total_data_train_sel.csv")
total_data_test_sel = pd.read_csv("/workspaces/machine-learning-boosting-algorithms/data/processed/total_data_test_sel.csv")
total_data_sel = pd.concat([total_data_train_sel, total_data_test_sel]).reset_index(drop = True)


X_train_sel = total_data_train_sel.drop(["Outcome"], axis = 1)
X_test_sel = total_data_test_sel.drop(["Outcome"], axis = 1)
y_train = total_data_train_sel["Outcome"]
y_test = total_data_test_sel["Outcome"]
```

```python

#Initialization and training of the model
from xgboost import XGBClassifier
model = XGBClassifier(random_state = 8)
model.fit(X_train_sel, y_train)

```

```python

#Model prediction
y_pred = model.predict(X_test_sel)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

The datasets with best performance are total_data_no_outliers_sel, total_data_no_outliers_norm_sel and total_data_no_outliers_scal_sel. We will take total_data_no_outliers_norm_sel for the optimization.

```python
#Model optimization

#Grid search
from sklearn.model_selection import GridSearchCV

# We define the parameters that we want to adjust by hand
hyperparams = {
    "n_estimators": [50, 80, 100],
    "learning_rate": [0.01, 0.05, 0.1],
    "gamma": [0, 0.5, 1, 5],
    "subsample": [0.5, 0.8, 1.0],
    "max_depth": [3, 4, 5, 6],
    "min_child_weight": [1, 5, 10],
    }

```

```python

# We initialize the grid
grid = GridSearchCV(model, hyperparams, scoring = "accuracy", cv = 10)
grid
```

```python

def warn(*args, **kwargs):
    pass
import warnings
warnings.warn = warn

grid.fit(X_train_norm_sel, y_train)

print(f"Best hyperparameters: {grid.best_params_}")
```

```python
model_grid = XGBClassifier(random_state = 8, gamma=5, learning_rate=0.1, max_depth=4, min_child_weight=1, n_estimators=50, subsample=0.8)
model_grid.fit(X_train_norm_sel, y_train)
y_pred = model_grid.predict(X_test_norm_sel)

grid_accuracy = accuracy_score(y_test, y_pred)
grid_accuracy

```

```python
from pickle import dump

dump(model_grid, open("../models/XGBClassifier_gamma-5_learningrate-0.1_maxdepth-4_minchildweight-1_estimators-50_subsample-0.8_8.sav", "wb"))
```

Comparing the results of training total_data_norm_sel with three models: Decision Tree, Random Forest and Boosting.

```python
#Decision Tree model with total_data_norm_sel
from sklearn.tree import DecisionTreeClassifier 
from pickle import load
load(open("../models/tree_classifier_crit-gini_maxdepth-5_minleaf-1_minsplit-2_8.sav", "rb"))

model_grid = DecisionTreeClassifier(criterion="gini", max_depth=5, min_samples_leaf=1, min_samples_split=2, random_state=8)
model_grid.fit(X_train_norm_sel, y_train)
y_pred = model_grid.predict(X_test_norm_sel)

grid_accuracy = accuracy_score(y_test, y_pred)
grid_accuracy
```

```python
#Random Forest model with total_data_norm_sel
from sklearn.ensemble import RandomForestClassifier
load(open("../models/RandomForestClassifier_estimators-100_criterion-gini_minsamplesleaf-2_minsamplessplit-6_8.sav", "rb"))

model_grid = RandomForestClassifier(criterion="gini", max_depth=5, min_samples_leaf=2, min_samples_split=6, n_estimators=100, random_state=8)
model_grid.fit(X_train_sel, y_train)
y_pred = model_grid.predict(X_test_sel)

grid_accuracy = accuracy_score(y_test, y_pred)
grid_accuracy
```

```python
#Boosting model with total_data_norm_sel
model_grid = XGBClassifier(random_state = 8, gamma=5, learning_rate=0.1, max_depth=4, min_child_weight=1, n_estimators=50, subsample=0.8)
model_grid.fit(X_train_norm_sel, y_train)
y_pred = model_grid.predict(X_test_norm_sel)

grid_accuracy = accuracy_score(y_test, y_pred)
grid_accuracy
```

Based on the results, the Random Forest model has been the one to perform best (even though in comparison with the boosting the difference is minimal) with this dataset.
